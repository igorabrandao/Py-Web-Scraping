```python

from urllib.request import urlopen
from urllib.error import HTTPError
from bs4 import BeautifulSoup

def getSoup(url):
    try:
        html = urlopen(url)
    except HTTPError as e:
        return None
    try:
        soup = BeautifulSoup(html.read())
    except AttributeError as e:
        return None
    return soup


def getTitle(url):
    try:
        html = urlopen(url)
    except HTTPError as e:
        return None
    try:
        soup = BeautifulSoup(html.read())
        title = soup.body.h1
    except AttributeError as e:
        return None
    return title

#title = getTitle("https://www.skoob.com.br/livro/similares/521364/edicao:528680")

#if title == None:
#    print("Title could not be found")
#else:
#    print(title)
    
url = "http://www.pythonscraping.com/pages/warandpeace.html"
soup = getSoup(url)

nameList = soup.find_all('span', {'class':'green'})
for name in nameList:
    print(name.get_text())
```

```python
# Vou fazer uma listagem de todos os links que aparecem na pagina "http://en.wikipedia.org/wiki/Kevin_Bacon"
soup = getSoup("http://en.wikipedia.org/wiki/Kevin_Bacon")
for link in soup.find_all("a"):
    if 'href' in link.attrs:
        print(link.attrs['href'])
```

```python
# Vou imprimir novamente a lista de links s√≥ que desta vez vou usar um filtro, onde apenas pegarei 
# links dentro da div body content, tamb√©m excluirei resultados que tenham ponto e virgula,  e por fim pegarei 
# apenas os links que come√ßam com /wiki/
for link in soup.find("div", {"id":"bodyContent"}).findAll("a",href=re.compile("^(/wiki/)((?!:).)*$")):
    if 'href' in link.attrs:
        print(link.attrs['href'])
```

```python
# VOu criar uma fun√ß√£o que retorna uma lista de links
def getLinks(soup):
    linkList = []
    for link in soup.find("div",{"id":"bodyContent"}).findAll("a",href=re.compile("^(/wiki/)((?!:).)*$")):
        if 'href' in link.attrs:
            linkList.append(link.attrs['href'])
    return linkList

# Vou testar agora o getLinks
listaDeLinks = getLinks(soup)
listaDeLinks
```

```python
# Vers√£o melhorada do getLink 
import datetime
import random

random.seed(datetime.datetime.now())

def getLinks(articleUrl):
    url = "http://en.wikipedia.org"+articleUrl
    soup = getSoup(url)
    return soup.find("div",{"id":"bodyContent"}).findAll("a",href=re.compile("^(/wiki/)((?!:).)*$"))

links = getLinks("/wiki/Kevin_Bacon")

while len(links) > 0:
    newArticle = links[random.randint(0, len(links)-1)].attrs["href"]
    print(newArticle)
    links = getLinks(newArticle)
```
